### 1st Momemt Business Decision (Measure of Central Tendency ) : mean(), median(), mode()

Mean : 

    which can be heavily influenced by extreme values (outliers) in the dataset, the median is resistant to outliers. 
    This makes it a more robust measure of central tendency in the presence of extreme values.

Median : The median represents the middle value of a dataset when it is ordered from smallest to largest. This makes it a good measure of central tendency when the data is skewed or not normally distributed.

```python
print(x[numeric_features].mean()) 
print("\n")

print(x[numeric_features].median())
print("\n")

print(x[numeric_features].mean())
```

### 2nd Moment Decision (Measure of Dispersion): Varience(), StandardDeavition(), Range()

Variance and standard deviation provide information about how spread out the values in a dataset are from the mean (average) value.

A high variance or standard deviation indicates that the data points are spread out over a wider range, while a low variance or standard deviation indicates that the data points are clustered closely around the mean.

```python
print(x[numeric_features].var())
print("\n")

print(x[numeric_features].std())
print("\n")
```

### 3rd Moment Decision (Measure of skewness (Measure of asymmetry in the data distribution)):skewness()

Skewness tells us whether the data is **symmetrically distributed or skewed to one side**. A skewness of **0** indicates a perfectly **symmetrical distribution**.

**If skewness is positive**, it means the data is skewed to the right **(long tail on the right side)**, indicating that the majority of the data points are on the left side with a few large values pulling the mean to the right.

**if skewness is negative**, it means the data is skewed to the left **(long tail on the left side)**.

```python
print(x[numeric_features].skew())
print("\n")
```

### 4th Moment Business Decision (kurtosis (Measure of peakedness) : It represents overall spread in the data): kurtosis()

Kurtosis is a statistical measure that **describes the shape, or peakedness,of the distribution of data relative to a normal distribution**

It tells you about the tails of your distribution and whether they are heavier or lighter than those of a normal distribution.

```python
print(x[numeric_features].kurt())
print("\n")
```

## Handling missing values :

```python
# Handling missing values for numerical data using mean or median imputation methods :SimpleImputer
```

```python
#Define pipeline for missing data if any
    
num_pipeline = Pipeline(steps = [('impute', SimpleImputer(strategy = 'mean'))])

num_pipeline
```

```python
preprocessor = ColumnTransformer(transformers = [('num', num_pipeline, numeric_features)])

preprocessor
```

```python
imputation = preprocessor.fit(x) # fix the data in preprocessor

imputation
```

```python
joblib.dump(imputation, 'meanimpute') # save the pipeline in imputation using joblib
```

```python
cleandata = pd.DataFrame(imputation.transform(x), columns = numeric_features) # Transform the imputation data 
```

## checking for missing values

```python
cleandata.isnull().sum() # checking for missing values
```

```python
# create pipeline using OneHotEncoder 

encoding_pipeline = Pipeline([('onehot', OneHotEncoder(drop = 'first'))]) # create pipeline in Encoding

encoding_pipeline
```

```python
preprocess_pipeline = ColumnTransformer([('categorical', encoding_pipeline, categorical_features)])

preprocess_pipeline
```

```python
#Fit the data in OneHotEncoding Pipeline

clean =  preprocess_pipeline.fit(x[categorical_features]) # fit the categorical data in Pipeline

clean
```

```python
# Save the Encoding Model

joblib.dump(clean, 'encoding') 
```

```python
#Transform the data usig OneHotEncoder

encode_data = pd.DataFrame(clean.transform(x[categorical_features]), columns = clean.get_feature_names_out())

encode_data
```

## Scaling :

### It involves transforming the features of a dataset onto a similar scale.

Data scaling, also known as feature scaling or normalization, is a crucial step in data preprocessing for machine learning models.

**There are various techniques for data scaling :**

**Min-Max Scaling (Normalization):** Scales features to a fixed range, usually between 0 and 1.
    
**Standardization:** Scales features to have a mean of 0 and a standard deviation of 1.

```python
# ## Scaling with MinMaxScaler

scale_pipeline = Pipeline([('scale', MinMaxScaler())])

scale_pipeline
```

```python
scale_columntransfer = ColumnTransformer([('scale', scale_pipeline, numeric_features)])

scale_columntransfer
```

```python
#fit the data in MinMaxScaler Pipeline

scale = scale_columntransfer.fit(cleandata1)

scale
```

```python
## Save minmax model

joblib.dump(scale, 'minmax')
```

```python
#Transform the data using MinMaxScaler

scaled_data = pd.DataFrame(scale.transform(cleandata1), columns = scale.get_feature_names_out())

scaled_data
```
